Fractal Generation 

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Define the fractal generation function
def mandelbrot(c, max_iter):
    z = 0
    n = 0
    while abs(z) <= 2 and n < max_iter:
        z = z**2 + c
        n += 1
    return n

# Set the parameters for the fractal
width, height = 500, 500
x_min, x_max = -2.5, 1.5
y_min, y_max = -2.0, 2.0
max_iter = 255

# Create the fractal image
fractal = np.zeros((height, width))
for y in range(height):
    for x in range(width):
        c = complex(x / width * (x_max - x_min) + x_min,
                    y / height * (y_max - y_min) + y_min)
        fractal[y, x] = mandelbrot(c, max_iter)

# Normalize the image data
fractal /= max_iter

# Create the plot
plt.figure(figsize=(10, 10))
plt.imshow(fractal, cmap='hot')
plt.title("Mandelbrot Set")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.axis("off")
plt.show()
```
