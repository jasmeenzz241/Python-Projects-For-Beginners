Simulating the Spread of a Disease

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
beta = 0.5     # transmission rate
gamma = 0.1    # recovery rate
N = 1000       # total population size.
I0 = 10        # initial number of infected individuals.
R0 = 0         # initial number of recovered individuals 
S0 = N -I0 -R0 # initial number of susceptible individuals

t = np.linspace(0,100)

S = S0 * np.exp(-beta*t)
I = I0 * np.exp((beta - gamma)* t)
R = R0 + gamma * I0 * (np.exp((beta - gamma)* t)-1)/(beta - gamma)

plt.plot(t, S, label = 'Susceptible' ,color='slategray')
plt.plot(t, I, label = 'Infected', color='rebeccapurple')
plt.plot(t, R, label = 'Recovered', color='sienna')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```
